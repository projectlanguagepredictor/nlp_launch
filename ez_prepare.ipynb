```python
import requests
from bs4 import BeautifulSoup

base_url = 'https://github.com/search?q=awesome%20lists&type=repositories'
page = 1

while True:
    url = f'{base_url}&p={page}'

    response = requests.get(url)

    soup = BeautifulSoup(response.text, 'html.parser')

    repo_link_elements = soup.select('.repo-list .v-align-middle')
#     print(repo_link_elements)

    for link_element in repo_link_elements:
        time.sleep(1)
       # html = link_element[]
        #print(link_element)
        print(page)
        try:
            href = link_element['href']
        except:
            print("no href")
        else:
            if href and href.startswith('/'):
                print(f'https://github.com{href}')
                print()

    next_button = soup.find('a', class_='next_page')

    if next_button is not None:
        page += 1
    else:
        break
```
